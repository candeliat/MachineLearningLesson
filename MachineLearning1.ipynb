```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Audi_A1_listings.csv')
```

```python
df.head(3)
```

```python
df=df.drop(columns=['index','href','MileageRank','PriceRank','PPYRank','Score','PPY']) # istemediğimiz sütunları siler
```

```python
df.info()
```

```python
df.head(3)
```

```python
df.columns=['yil','kasa','mil','motor','ps','vites','yakit','sahip','fiyat']
```

```python
df.head(3)
```

```python
df['motor']=df['motor'].str.replace('L','') #veri içindeki istemediğimiz değerleri siler
```

```python
df['motor']=pd.to_numeric(df['motor']) #verinin tipini değiştirir floata
```

```python
df.head(3)
```

```python
df=pd.get_dummies(df,columns=['kasa','vites','yakit'],drop_first=True) #girdiğimiz değerleri boolean değere değiştirir
```

```python
y=df['fiyat']
x=df.drop('fiyat',axis=1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=22)
```

```python
lm = LinearRegression()
model=lm.fit(x_train,y_train)
model.score(x_test,y_test)
```
