```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,mean_absolute_error,mean_absolute_percentage_error
```

```python
df= pd.read_csv('insurance.csv')
df.head()
```

```python
df=pd.get_dummies(df,columns=['sex','smoker','region'],drop_first=True)
```

```python
df.head(3)
```

```python
y=df['charges']
x=df.drop('charges',axis=1)
```

```python
lm = LinearRegression()
model=lm.fit(x,y)
```

```python
model.score(x,y)
```

```python
model.predict([[19,26,0,1,1,0,0,1]])
```

```python
df_hata = pd.DataFrame()
```

```python
df_hata['y']=y
```

```python
y_tahmin=model.predict(x)
```

```python
df_hata['tahmin']=y_tahmin
```

```python
df_hata
```

```python
df_hata['hata']=df_hata['y']-df_hata['tahmin']
```

```python
df_hata.head(10)
```

```python
df_hata['hata_kare']=df_hata['hata']**2
```

```python
df_hata.head(10)
```

```python
df_hata['abs_hata']=np.abs(df_hata['hata']) #MAE mutlak değer içine alıyoruz
```

```python
df_hata.head()
```

```python
df_hata['percent_hata']=np.abs(df_hata['hata'])/df_hata['y'] #MAPE Yüzdelik hataları alıyor
```

```python
df_hata.head()
```

```python
df_hata.mean()
```

```python
mean_squared_error(df_hata['y'],df_hata['tahmin'])
```

```python
mean_absolute_error(df_hata['y'],df_hata['tahmin'])
```

```python
mean_absolute_percentage_error(df_hata['y'],df_hata['tahmin'])
```
